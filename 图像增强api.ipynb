```python
# resize 缩放
# crop 裁剪
# flip 翻转
# brightess & contrast  改变光照 & 对比度

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
name = 'test.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
sess = tf.Session()
img_decoded_val = sess.run(img_decoded)
print(img_decoded_val.shape)
plt.imshow(img_decoded_val)
```

```python
# resize
# tf.image.resize_area
# tf.image.resize_bicubic
# tf.image.resize_nearest_neighbor

name = 'test.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1,512, 512, 4 ])

img_resize = tf.image.resize_bicubic(img_decoded, [1024, 1024])

sess = tf.Session()
img_decoded_val = sess.run(img_resize)
img_decoded_val = img_decoded_val.reshape((1024, 1024, 4))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)
plt.imshow(img_decoded_val)
```

```python
# crop
# tf.image.pad_to_bounding_box 用黑边填充边框
# tf.image.crop_to_bounding_box 裁剪边框
# tf.random_crop

name = 'test.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1,512, 512, 4 ])

img_padded = tf.image.pad_to_bounding_box(img_decoded, 20, 100, 700, 700)

sess = tf.Session()
img_decoded_val = sess.run(img_padded)
img_decoded_val = img_decoded_val.reshape((700, 700, 4))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)
plt.imshow(img_decoded_val)
```

```python
# flip
# tf.image.flip_up_down
# tf.image.flip_left_right
# tf.image.random_flip_up_down
# tf.image.random_flip_left_right

name = 'test.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1,512, 512, 4 ])

img_flipped = tf.image.flip_up_down(img_decoded)
sess = tf.Session()
img_decoded_val = sess.run(img_flipped)
img_decoded_val = img_decoded_val.reshape((512, 512, 4))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)
plt.imshow(img_decoded_val)
```

```python
# brightness
# tf.image.adjust_brightness
# tf.image.random_brightness
# tf.image.adjust_contrast
# tf.image.random_contrast

name = 'test.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
img_decoded = tf.reshape(img_decoded, [1,512, 512, 4 ])

img_brightness = tf.image.adjust_brightness(img_decoded, -0.5)
sess = tf.Session()
img_decoded_val = sess.run(img_brightness)
img_decoded_val = img_decoded_val.reshape((512, 512, 4))
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)
plt.imshow(img_decoded_val)
```
